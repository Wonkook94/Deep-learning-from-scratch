```python
import numpy as np
import matplotlib.pyplot as plt
```

### Input X

```python
x = np.arange(-5.0, 5.0, 0.1)
# [-5.0, 5.0) 구간에서 step(0.1) 만큼 숫자를 array 형태로 반환
```

### 3.2.3 Step Function 

```python
def step_function(x):
    return np.array(x > 0, dtype=np.int32)

y = step_function(x)
plt.plot(x,y,label='step function')
plt.ylim(-0.1, 1.1)    # y축 범위 설정
plt.show()
```

### 3.2.4 Sigmoid Function

```python
def sigmoid(x):
    return 1 / (1+np.exp(-x))

y = sigmoid(x)
plt.plot(x,y,label='sigmoid function')
plt.ylim(-0.1,1.1)
plt.show()
```

### 3.2.7 ReLU Function

```python
def relu(x):
    return np.maximum(0,x)   # maximum은 입력 중 큰 입력을 반환한다.

y = relu(x)
plt.plot(x,y,label='relu function')
plt.ylim(-2.0,5.0)
plt.show()
```

### Identity Function

```python
def identity_function(x):
    return x

y = identity_function(x)
plt.plot(x,y,label='identity function')
plt.ylim(-5.0,5.0)
plt.show()
```

### 3.4.3 신경망 구현하기

```python
def init_network():
    network={}  # 딕셔너리 자료형
    network['W1'] = np.array([[0.1,0.3,0.5],[0.2,0.4,0.6]])
    network['b1'] = np.array([0.1,0.2,0.3])
    network['W2'] = np.array([[0.1,0.4],[0.2,0.5],[0.3,0.6]])
    network['b2'] = np.array([0.1,0.2])
    network['W3'] = np.array([[0.1,0.3],[0.2,0.4]])
    network['b3'] = np.array([0.1,0.2])
    
    return network
```

```python
def forward(network, x):
    W1, W2, W3 = network['W1'], network['W2'], network['W3']
    b1, b2, b3 = network['b1'], network['b2'], network['b3']

    a1 = np.dot(x,W1) + b1
    z1 = sigmoid(a1)  # 활성화 함수로는 sigmoid
    a2 = np.dot(z1, W2) + b2
    z2 = sigmoid(a2)   # 활성화 함수로는 sigmoid
    a3 = np.dot(z2, W3) + b3
    y = identity_function(a3)  # 활성화 함수로는 identity

    return y
```

```python
x = np.array([1.0, 0.5])  # Input x

network = init_network()
y = forward(network,x)
print(y)
```

### 3.5.1 Softmax Function

```python
def softmax(x):
    '''
    exp_x = np.exp(x)
    sum_exp_x = np.sum(exp_x)
    y = exp_x / sum_exp_x
    
    정의대로 구현한 식이지만, overflow 발생한다. 
    '''
    c = np.max(x)
    exp_x = np.exp(x-c)   # overflow를 방지하자.
    sum_exp_x = np.sum(exp_x)
    y = exp_x / sum_exp_x
    
    return y
```

### 3.6.2 Neural Network Infer

```python
def get_data():
    (x_train,y_train), (x_test,y_test) = load_mnist(flatten=True, normalize=False)

    return x_test,y_test
```

```python
def init_network():
    with open("sample_weight.pkl",'rb') as f:
        network = pickle.load(f)

    return network
```

```python
def predict(network, x):
    W1, W2, W3 = network['W1'], network['W2'], network['W3']
    b1, b2, b3 = network['b1'], network['b2'], network['b3']

    a1 = np.dot(x, W1) + b1
    z1 = sigmoid(a1)
    a2 = np.dot(z1, W2) + b2
    z2 = sigmoid(a2)
    a3 = np.dot(z2, W3) + b3
    y = softmax(a3)

    return y
```

```python
x, t = get_data()
network = init_network()
```

```python
accuracy_cnt = 0

for i in range(len(x)):
    y = predict(network,x[i])
    p = np.argmax(y)
    if p == t[i]:
        accuracy_cnt += 1
        
print("Accuracy: "+str(float(accuracy_cnt)/len(x)))
```

### 3.6.3 Batch 

```python
x, t = get_data()
network = init_network()

batch_size = 100 # 배치 크기
accuracy_cnt = 0
for i in range(0,len(x),batch_size):
    x_batch = x[i:i+batch_size]
    y_batch = predict(network,x_batch)
    p = np.argmax(y_batch,axis=1)
    accuracy_cnt += np.sum(p==t[i:i+batch_size])


print("Accuracy: "+str(float(accuracy_cnt)/len(x)))
```
